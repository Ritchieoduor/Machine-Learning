# KNN (K nearest neighbours)

```python
# importing necessary libraries 

import numpy as np
import pandas as pd
from sklearn import neighbors,metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
# importing the car data

columns = ["buying","maint","doors","person","lug_boot","safety","class"]
car_data = pd.read_csv("car.data")
car_data.columns = columns # changing the columns names
car_data.head()
```

```python
#X = car_data[car_data.columns[:-1]]  #all the columns except the last one
X = car_data[["buying","maint","safety"]].values # converting it to an array using the .values function
y = car_data[["class"]] #grabbing only the last column

X
```

```python
y.head()
```

```python
X.ndim # a 2 dimensional array
#X.shape
```

```python
X
```

### Data Cleaning

```python
Le = LabelEncoder() # converting the values to numerics using labelencoder function

for i in range(len(X[0])):
    X[:, i] = Le.fit_transform(X[:, i])
    
print(X)


# vhigh is mapped to value 3 ,med is mapped to 2,low mapped to 1 and high to 0
```

```python
# converting y using mapping
#this is another way of converting the values of a column
y_mapping = {"unacc":0,
            "acc":1,
            "good":2,
            "vgood":3,}
y["class"] = y["class"].map(y_mapping)

y = np.array(y)
y

```

```python
# creating the knn object

knn = neighbors.KNeighborsClassifier(n_neighbors = 25,weights ="uniform")

#n_neighbors is the the number of neighbors to be compaired to the the object provided
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# Training the model

knn.fit(X_train,y_train)
```

```python
# prediction and accuracy of the model

prediction = knn.predict(X_test)
accuracy = metrics.accuracy_score(y_test, prediction)
```

```python
print(f"The prediction of the model is:\n{prediction}")
```

```python
print("The accuracy of the model is " + str(accuracy*100) + "%")
```

### Digit recognition using SVM

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
X1 = digits.data
y = digits.target

print(X1)
print(X1.shape)
print(digits)
```

```python
n = 500 # number of sample out of the sample total 1797

plt.gray() #gray scale
plt.matshow(digits.images[n])
plt.show()

digits.images[n]
```

```python
digit_img_len = len(digits.images)
print(digit_img_len)


# segregating the data
X = digits.images.reshape(digit_img_len,-1)
print(X.shape)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state=0)
print(X_train.shape,X_test.shape,sep="\n")



```

```python
model = SVC(gamma = 0.001)
model.fit(X_train,y_train)

```

```python
pred = model.predict(X_test)
print("prediction:\n",pred)
```

```python
acc = accuracy_score(y_test,pred)
print("accuracy is:\n"+str(acc*100)+"%")

```

### Tuning the model

```python
#setting kernel to linear

model1 = SVC(kernel = "linear")
model1.fit(X_train,y_train)
pred1 = model1.predict(X_test)
print("model accuracy is ",accuracy_score(y_test,pred1)*100)
```

```python
model3 = SVC(gamma = 0.001,C= 0.1)
model3.fit(X_train,y_train)
pred3 = model3.predict(X_test)
print("model accuracy is ",accuracy_score(y_test,pred3)*100)
```

#### first model is the best because of its high accuracy

## DECISION TREE

```python
# import necessary libraries
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
X = iris.data
y= iris.target
print(iris.target_names)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25)
X_train.shape
```

```python
# checking for the max_depth that provides the best accuracy
acc_list = []
for i in range(1,6):
    ds = DecisionTreeClassifier(max_depth = i,random_state= 0)
    
    ds.fit(X_train,y_train)
    prediction = ds.predict(X_test)
    acc = (accuracy_score(y_test,prediction))*100
    acc_list.append(acc)
    
    
```

```python
print(acc_list)
```

```python
# plot the graph of max_depth against the accuracy scores
plt.plot([i for i in range(1,6)],acc_list,"ro--",label = "accuracy")
plt.title("accuracy score graph")
plt.xlabel("max_depth")
plt.ylabel("accuracy_score")
plt.legend()
plt.show()

#accuracy start to be better at a max depth of 2 as shown by the plot
          
```

```python
d1 = DecisionTreeClassifier(criterion ="entropy", max_depth = 4,random_state=0)

```

```python
d1.fit(X_train,y_train)
pred = d1.predict(X_test)
acc = accuracy_score(y_test,pred)

print(pred)
print(f"accuracy score of the model is {acc*100:.2f}%")
```
